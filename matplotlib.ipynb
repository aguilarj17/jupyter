```python
import matplotlib.pyplot as plt
import pandas as pd



t001 = pd.read_csv('wktime.T001.his', delimiter=' ', header=None) #Import the file 
t004 = pd.read_csv('wktime.T004.his', delimiter=' ', header=None) 

#Drawing the first line of  graph
plt.plot(t001[0], t001[1], color='b', label='Time 1') 
#Drawing the second graph with the values of the file
plt.plot(t004[0], t004[1], color='r', label='Time 2') #Drawing the second plot graph

#components of the graph 
plt.title("Relation between wealth and probability")
plt.xlabel("Wealth")       #name of the axis x
plt.ylabel("Probability")  #name of the axis y
plt.xscale("log")                      # axis X scale
plt.yscale("log")                      #Axis Y scale
plt.xlim((0.000001, 100.0))            # axis X limits
plt.ylim((0.000001, 100000.0))         #axis y limits
plt.tight_layout()






plt.show()
```
